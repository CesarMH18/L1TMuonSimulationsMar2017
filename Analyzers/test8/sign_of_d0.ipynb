# Sign of transverse impact parameter d<sub>0</sub>

1. We define the transverse impact parameter d<sub>0</sub> as the distance of closest approach of the track __to the z-axis__.
  - If defined w.r.t the beam axis, we need correction for the beam movement.
  - If defined w.r.t the primary vertex, which is a 3D point, the equation becomes far more complicated.

2. We assume that the magnetic field is pointing along the positive z-axis. A track with negative charge goes counter-clockwise, i.e. it has increasing &#981;(r). 

3. We also define the radius of curvature R as:

\begin{align}
R = -\frac{p_{\mathrm{T}}}{0.003 q B}\,[\mathrm{cm}]
\end{align}

where q is the charge of the track, and B is the magnetic field strength in Tesla. R is positive if the track goes counter-clockwise.

4. The sign of d<sub>0</sub> needs to satisfy the following two requirements:

  - The sign should agree with the sign of $(\vec{r} \times \vec{p})_{z}$ at the point of closest approach, where $\vec{r}$ is the point of closest approach from the origin. This also means that the sign of d<sub>0</sub> is positive if the direction of the track at d<sub>0</sub> is counter-clockwise w.r.t. the origin.
  - The d<sub>0</sub> is continuous such that the sign changes from positive to negative (or vice versa) at d<sub>0</sub> = 0.

5. For a straight track with infinite p<sub>T</sub>, the d<sub>0</sub> with the correct sign can be obtained from the following equation:

\begin{align}
d_{0} = x_{v} \sin{\phi} - y_{v} \cos{\phi}
\end{align}

where (x<sub>v</sub>, y<sub>v</sub>) are the vertex coordinates and &#981; is the azimuthal angle of the track $\vec{p}_{\mathrm{T}}$ at the vertex. If we express (x<sub>v</sub>, y<sub>v</sub>) in polar coordinates (R<sub>v</sub>, &#981;<sub>v</sub>), we get the equation for the sign of d<sub>0</sub> for a straight track:

\begin{align}
\operatorname{sign}(d_{0}) &= \operatorname{sign}(\cos{\phi_{v}} \sin{\phi} - \sin{\phi_{v}} \cos{\phi}) \\
      &= \operatorname{sign}(\sin(\phi - \phi_{v}))
\end{align}

6. For a curved track, it's more complicated. The absolute value of d<sub>0</sub> can be obtained from the following equation:

\begin{align}
\left|d_{0}\right| = \left|{|R| - \sqrt{{x_{c}}^2 + {y_{c}}^2}}\right|
%d_{0} = R - \sqrt{{x_{c}}^2 + {y_{c}}^2}
\end{align}

where (x<sub>c</sub>, y<sub>c</sub>) are the coordinates of the center of the helix in the x-y plane. They are related to (x<sub>v</sub>, y<sub>v</sub>) by the following equations:

\begin{align}
x_{c} &= x_{v} - R \sin{\phi} \\
y_{c} &= y_{v} + R \cos{\phi}
\end{align}

7. Finally, we find the sign of d<sub>0</sub>, which depends on (x<sub>v</sub>, y<sub>v</sub>), &#981; and R.

  - Find the sign of d<sub>0</sub> assuming a straight track, called S<sub>R&infin;</sub>.
  - Find the sign of R. If the sign of R is opposite to S<sub>R&infin;</sub>, the sign of d<sub>0</sub> will be S<sub>R&infin;</sub>; if the sign of R is the same as S<sub>R&infin;</sub>, the sign of d<sub>0</sub> will change as d<sub>0</sub> goes through zero.
  - Find the R for which d<sub>0</sub> goes through zero, called R\*. If |R| of the track is larger than |R\*|, the sign of d<sub>0</sub> will be S<sub>R&infin;</sub>; else, the sign of d<sub>0</sub> will be opposite to S<sub>R&infin;</sub>.
  
R\* can be obtained from the following equation:

\begin{align}
R^* = \frac{\sqrt{{x_{v}}^2 + {y_{v}}^2}}{2 \sin(\phi - \phi_{v})}
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

step = 1.0
x = np.arange(-120+1, 120, step)
y = np.arange(-120+1, 120, step)
X, Y = np.meshgrid(x, y)

phi = 135
phi_rad = np.deg2rad(phi)
Z = X * np.sin(phi_rad) - Y * np.cos(phi_rad)
Z = np.sign(Z)
```

```python
fig, ax = plt.subplots(figsize=(8,6.5))
CS = ax.contourf(X, Y, Z, cmap=plt.cm.coolwarm)
cbar = fig.colorbar(CS)
ax.axhline(y=0, alpha=0.5, color='white')
ax.axvline(x=0, alpha=0.5, color='white')
ax.set_title(r'Sign of $d_0$ for a straight track with $\phi = %i^{\circ}$' % phi, fontsize=16)
ax.set_xlabel(r'$x_v$ [cm]', fontsize=14)
ax.set_ylabel(r'$y_v$ [cm]', fontsize=14)

plt.show()
```

### Visualization

```python
xv, yv = 50, 50
phi = 135
invPt = -0.5

phi_rad = np.deg2rad(phi)
R = -(1.0/(0.003 * 3.8 * invPt))
xc, yc = xv - R * np.sin(phi_rad), yv + R * np.cos(phi_rad)

R1 = -1 * R
xc1, yc1 = xv - R1 * np.sin(phi_rad), yv + R1 * np.cos(phi_rad)

R2 = np.sqrt(xv*xv + yv*yv) / (2 * np.sin(phi_rad - np.arctan2(yv, xv)) + 1e-7)
xc2, yc2 = xv - R2 * np.sin(phi_rad), yv + R2 * np.cos(phi_rad)

simC = 1./(2*R) 
simCstar = np.sin(phi_rad - np.arctan2(yv, xv)) / np.sqrt(xv*xv + yv*yv)
d0 = np.abs(np.abs(R) - np.sqrt(xc*xc + yc*yc))
d0_sign = np.sign(simCstar)
if ((simCstar < 0) and (simC < simCstar)) or ((simCstar > 0) and (simC > simCstar)):
  d0_sign *= -1
d0 *= d0_sign

print R, R1, R2, d0
```

```python
from matplotlib.patches import Circle
vertex = Circle((xv, yv), 3, color='k')
center = Circle((xc, yc), 3, color='r')
circ = Circle((xc, yc), R, color='r', fill=False)
center1 = Circle((xc1, yc1), 3, color='b')
circ1 = Circle((xc1, yc1), R1, color='b', fill=False)
center2 = Circle((xc2, yc2), 3, color='g')
circ2 = Circle((xc2, yc2), R2, color='g', fill=False)

fig, ax = plt.subplots(figsize=(8,8))
ax.add_artist(vertex)
ax.add_artist(center)
ax.add_artist(circ)
ax.add_artist(center1)
ax.add_artist(circ1)
ax.add_artist(center2)
ax.add_artist(circ2)
ax.axhline(y=0, alpha=0.5, color='black')
ax.axvline(x=0, alpha=0.5, color='black')
ax.set_xlim((-300, 300))
ax.set_ylim((-300, 300))

plt.show()
```
